```python
import numpy
from scipy.optimize import least_squares
import matplotlib.pyplot as plt

def fun(a, x, y):
  return numpy.polyval(a, x) - y

def fx(x):
  return numpy.exp(numpy.cos(x))

x = numpy.array([x/10 for x in range(1, 11, 1)])
print(x)

y = numpy.array([fx(i) for i in x])
print(y)

a0 = numpy.array([1, 1])

res_lsq = least_squares(fun, x0=a0, args=(x, y))

print("a0 = %.2f, a1 = %.2f" % tuple(res_lsq.x))

x_p = numpy.linspace(min(x), max(x), 100)
y_p = numpy.polyval(res_lsq.x, x_p)

plt.plot(x, y, 'o', color="red")
plt.plot(x_p, y_p, color='green')
plt.title("Least squares linear method")
plt.grid(True)
plt.show()
```

```python
x = numpy.array([x/10 for x in range(1, 11, 0)])
y = numpy.array([fx(i) for i in x]) # Оцінюємо значення функції func для кожного x

print(x)
print(y)

def fun(a, x, y):
  return a[0] + a[1] * x + a[2] * x**2 - y

a0 = numpy.array([1, 1, 1])
res_lsq = least_squares(fun, x0=a0, args=(x, y))

print("a0 = %.2f, a1 = %.2f, a2 = %.2f" % tuple(res_lsq.x))

f = lambda x: sum([u * v for u, v in zip(res_lsq.x, [1, x, x**2])])
x_p = numpy.linspace(min(x), max(x), 20)
y_p = f(x_p)

plt.plot(x, y, 'o', color="red")
plt.plot(x_p, y_p, color='green')
plt.title("Least squares parabola")
plt.grid(True)
plt.show()
```
