```python
import numpy
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

table_x = numpy.array([-1., 0., 1., 5.], dtype = float)
table_fx = numpy.array([11., 4., 7., -1.], dtype = float)

given_x = numpy.array([-0.5, 2., 3., 4.5], dtype = float)
given_fx = numpy.array([], dtype = float)

def lgr_interp(x, y, x_assingment):

    n = len(x)
    p = numpy.zeros(n)

    for i in range(n):
        p_i = 1
        for j in range(n):
            if i != j:
                p_i *= (x_assingment - x[j])/(x[i] - x[j])
        p[i] = p_i
    #print(numpy.dot(y, p))
    return numpy.array(numpy.dot(y, p))
#print(type(lgr_interp(table_x, table_fx, given_x[1])))
for i in given_x:
  result = numpy.append(given_fx, lgr_interp(table_x, table_fx, i))
  print(f"Значення функції у точці {i} = {result[0]}")
```

```python
xnew=numpy.linspace(numpy.min(table_x),numpy.max(table_x),100)
fxnew=[lgr_interp(table_x,table_fx,i) for i in xnew]
plt.plot(table_x,table_fx,'o',xnew,fxnew, c='g', mfc='r', mec='r')
plt.title('Lagrange Polynomial')
plt.grid(True)
plt.show()
```

```python
f = lagrange(table_x, table_fx)
fig = plt.figure(figsize = (8,5))
plt.plot(xnew, f(xnew), 'g', table_x, table_fx, 'ro')
plt.title('Lagrange Polynomial Check')
plt.grid()
plt.xlabel('table_x')
plt.ylabel('table_fx')
plt.show()
```
